## 3: DATA 

```python
def get_football_data():
    print("Downloading Premier League data...")
    season_urls = {
        '2022-23': 'https://www.football-data.co.uk/mmz4281/2223/E0.csv',
        '2023-24': 'https://www.football-data.co.uk/mmz4281/2324/E0.csv',
        '2024-25': 'https://www.football-data.co.uk/mmz4281/2425/E0.csv'
    }

    all_matches = []
    for season, url in season_urls.items():
        df = pd.read_csv(url)
        df["Season"] = season
        all_matches.append(df)

    match_data = pd.concat(all_matches, ignore_index=True)
    return match_data


df = get_football_data()
df.head()
```

### **Data Summary**

The data is derived from `https://www.football-data.co.uk`. It contains **1140** observations and **133** attributes of **match metadata** and **bookmaker odds** spanning the last three seasons. Most of bookmaker odds data are null. There are no duplicates in the data

I will retain only the **match metadata**, **scoreline information**, and **performance statistics** because these variables describe the game itself without relying on betting market data. All other columns in the dataset are **bookmaker odds**, which pose two main problems for my goal of predicting the **2025–26 Premier League winner**:

#### 1. Data Leakage
- Odds are set just before each match and already incorporate knowledge about team strength, injuries, and other factors unavailable at the start of the season.  
- Using them would effectively feed my model the outcome probabilities determined by bookmakers, making predictions **unrealistically accurate** for past seasons but **unusable before future games**.

#### 2. Future Applicability
- For the 2025–26 season, **pre-match odds for every fixture** are not available in advance.  
- Relying on them would prevent me from making an accurate prediction. Retaining only the **non-odds columns** — i.e., **goals, shots, fouls, cards**, and other **in-game statistics** I will ensure that the model accuracy is optimum.

## 5: **INFERENTIAL ANALYSIS**

### 5.1 Testing if the average performance differs home vs away

#### 5.1.1 Formulating Hypothesis
Null hypothesis (H₀): The mean performance shots or goals is the same for home and away.

Alternative hypothesis (H₁): The mean performance is different between home and away.

**Shots:**

H₀: mean(HS) = mean(AS)

H₁: mean(HS) ≠ mean(AS)

**Goals:**

H₀: mean(FTHG) = mean(FTAG)

H₁: mean(FTHG) ≠ mean(FTAG)

#### 5.1.2 Conducting the T test

```python
from scipy.stats import ttest_ind
home_shots = data['HS']
away_shots = data['AS']

mean_home = np.mean(home_shots)
mean_away = np.mean(away_shots)
std_home = np.std(home_shots, ddof=1)
std_away = np.std(away_shots, ddof=1)
mean_diff = mean_home - mean_away

t_stat, p_value = ttest_ind(home_shots, away_shots, equal_var = False)
print(f"T-statistic: {t_stat}, p-value: {p_value}")
print("\n")
if p_value < 0.05:
    print("We reject the null hypothesis and postulate that at 0.05 statistical significance, the mean for home teams shots is statistically different with mean for away teams highlighting possible home advantage")

else:
    print("We fail to reject the null hypothesis and postulate that at 0.05 statistical significance, there is no enough evidence that the mean for home teams shots is statistically different with mean for away teams highlighting possible home advantage")
```

```python
#  Cohen's d (pooled SD)
pooled_sd = np.sqrt(((len(home_shots)-1)*std_home**2 + (len(away_shots)-1)*std_away**2) / (len(home_shots)+len(away_shots)-2))
cohens_d = mean_diff / pooled_sd
print(f"Cohen's d: {cohens_d:.3f}")
```

```python
home_goals = data['FTHG']
away_goals = data['FTAG']

mean_home = np.mean(home_goals)
mean_away = np.mean(away_goals)
std_home = np.std(home_goals, ddof=1)
std_away = np.std(away_goals, ddof=1)
mean_diff = mean_home - mean_away

t_stat, p_value = ttest_ind(home_goals, away_shots, equal_var=False)
print(f"T-statistic: {t_stat}, p-value: {p_value}")

print("\n")
if p_value < 0.05:
    print("We reject the null hypothesis and postulate that at 0.05 statistical significance, the mean for home team goals is statistically different with mean for away teams highlighting possible home advantage")
else:
    print("We fail to reject the null hypothesis and postulate that at 0.05 statistical significance, there is no enough evidence that the mean for home teams shots is statistically different with mean for away teams highlighting possible home advantage")
```

```python
#  Cohen's d (pooled SD)
pooled_sd = np.sqrt(((len(home_goals)-1)*std_home**2 + (len(away_goals)-1)*std_away**2) / (len(home_goals)+len(away_goals)-2))
cohens_d = mean_diff / pooled_sd
print(f"Cohen's d: {cohens_d:.3f}")
```

#### 5.1.3 Drawing Inference from the Test

**The p_value for both goals and shots is less than 0.05 highlighting that there is statistical significant difference between home teams and away teams performance over the period which can be attributed to home advantage i.e. more fans and so on.**

**However, applying Cohen's d measure, the difference is very small.**

### 5.2 Testing if the average performance differs home vs away in the same games

#### 5.2.1 Formulating Hypothesis
Null hypothesis (H₀): The mean performance shots or goals is the same for home and away in each game.

Alternative hypothesis (H₁): The mean performance is different between home and away in each game.

**Shots:**

H₀: mean(HS) = mean(AS)

H₁: mean(HS) ≠ mean(AS)

**Goals:**

H₀: mean(FTHG) = mean(FTAG)

H₁: mean(FTHG) ≠ mean(FTAG)

#### 5.2.2 Conducting the T test

```python
# Differences per match
Dg = data['FTHG'] - data['FTAG']   
Ds = data['HS']   - data['AS']     
```

```python
#checking if the difference has normality
from scipy.stats import shapiro
print(f"Goals Difference: {shapiro(Dg)}")
print(f"Shots Difference: {shapiro(Ds)}")
```

**The Shapiro-Wilk test results indicate that the **goals difference (Dg)** has a p-value of approximately 9.7716275446625e-14, which is far below the 0.05 significance threshold. This leads us to reject the null hypothesis of normality, meaning the data is not normally distributed; therefore, the **Wilcoxon signed-rank test** is more appropriate than the paired t-test for goals difference. In contrast, the **shots difference (Ds)** has a p-value of approximately 0.0713, which is greater than 0.05. We fail to reject the null hypothesis of normality in this case, suggesting the data is roughly normal, and the **paired t-test** is valid for comparing shots difference.**


```python
from scipy.stats import ttest_rel, wilcoxon

# Paired t-tests (two-sided) shots
t_shots, p_shots = ttest_rel(data['HS'],   data['AS'],   alternative='two-sided')
print(f"Paired t-test (Shots): t={t_shots:.3f}, p={p_shots:.3e}")

# Wilcoxon (Goals)
t_goals, p_goals = wilcoxon(data['FTHG'], data['FTAG'], alternative='two-sided')
print(f"Wilcoxon (Goals): t={t_goals:.3f}, p={p_goals:.3e}")
```

#### 5.2.3 Drawing Inference from the Paired Samples Analysis

**We compared home and away team performance using two statistical tests.** 
- `**Paired t-test (Shots):**` There was a statistically significant difference in the number of shots between home and away teams, *t*(n)=8.972, *p* < 0.05, indicating that home teams generally took more shots.  
- `**Wilcoxon Signed-Rank Test (Goals):**` A significant difference was also found in goals scored, *W* = 158,295, *p* < 0.05, suggesting home teams tended to score more goals.  

**These results support the hypothesis that home advantage plays a role in both shot attempts and goal scoring.**


### 5.3 Testing if Performance varies across 3+ groups

#### 5.3.1 Formulating Hypothesis

Null (H₀): The mean goals are the same across all referees.

Alternative (H₁): At least one referee’s mean goals differ from the others.


#### 5.3.2 Conducting the One_Way Anova

```python
from scipy.stats import f_oneway

groups = [data.loc[data['Referee'] == ref, 'FTHG'] for ref in data['Referee'].unique()]
f_stat, p_val = f_oneway(*groups)
print(f"ANOVA: F={f_stat:.3f}, p={p_val:.3e}")
```

```python
if p_val < 0.05:
    print("Reject the null hypothesis, at 0.05 significance level there is enough evidence to suggest that not all mean goals are same across all the referees")
else:
    print("We fail to reject the null hypothesis, at 0.05 significance level there is no enough evidence to suggest that all mean goals are not same across all the referees")
```

```python
#Trying the same for away goals
groups2 = [data.loc[data['Referee'] == ref, 'FTAG'] for ref in data['Referee'].unique()]
f_stat, p_val = f_oneway(*groups2)
print(f"ANOVA: F={f_stat:.3f}, p={p_val:.3e}")
```

```python
if p_val < 0.05:
    print("Reject the null hypothesis, at 0.05 significance level there is enough evidence to suggest that not all mean goals are same across all the referees")
else:
    print("We fail to reject the null hypothesis, at 0.05 significance level there is no enough evidence to suggest that all mean goals are not same across all the referees")
```

#### 5.3.3 Drawing Inference from the ANOVA Test


**A one-way ANOVA was conducted to test whether the mean number of goals differed across referees. The results indicated that there was **no statistically significant difference** in goals between referees (*p=8.182e-01* & *p=8.102e-02* are > 0.05). Therefore, we fail to reject the null hypothesis and conclude that, at the 0.05 significance level, there is not enough evidence to suggest that the mean goals vary across referees. This implies that referee identity does not appear to influence the number of goals scored in the matches analyzed.**


### 5.4 Chi-Square Test of Independence

#### 5.4.1 Formulating Hypothesis

Null (H₀): Match result is independent of referee.

Alternative (H₁): Match result depends on referee.

#### 5.4.2 Performing the Test

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data['Referee'], data['FTR'])
chi2, p_val, dof, expected = chi2_contingency(contingency_table)
print(f"Chi2={chi2:.3f}, p={p_val:.3e}, dof={dof}")
```

```python
#Interpretation of the results
if p_val < 0.05:
    print("Reject the null hypothesis: at the 0.05 significance level, there is a significant association between referee and match result, suggesting some referees may influence outcomes.")
else:
    print("Fail to reject the null hypothesis: at the 0.05 significance level, there is no significant association between referee and match result, suggesting outcomes appear independent of referee.")
```

#### 5.4.3 Drawing Inference from the Chi_Test

A Chi-Square test of independence was conducted to examine whether match result (H/D/A) is associated with referee. The results indicated that there was **no statistically significant association** between referee and match result (*Chi2* = 62.009, *p* = 0.5473, *dof* = 64). Therefore, we fail to reject the null hypothesis and conclude that, at the 0.05 significance level, there is no evidence to suggest that match outcomes depend on referee. This implies that referee identity does not appear to influence the results of the matches analyzed.


### 5.5 Pearson Correlation Test

#### 5.5.1 Formulating Hypothesis

Let’s say we want to test if shots (HS/AS) are correlated with goals (FTHG/FTAG):

Null (H₀): There is no correlation between the two variables (r=0)

Alternative (H₁): There is a correlation (r≠=0)

#### 5.5.2 Performing the Test

```python
from scipy.stats import pearsonr, spearmanr
x = data['HS'] + data['AS']
y = data['FTHG'] + data['FTAG']

r_pearson, p_pearson = pearsonr(x, y)
print(f"Pearson correlation: r={r_pearson:.3f}, p={p_pearson:.3e}")

r_spearman, p_spearman = spearmanr(x, y)
print(f"Spearman correlation: r={r_spearman:.3f}, p={p_spearman:.3e}")
```

#### 5.5.3 Drawing Inference from Correlation Tests

Pearson and Spearman correlation tests were conducted to examine the relationship between shots and goals. The Pearson correlation coefficient was *r* = 0.231, *p* < 0.05, and the Spearman correlation coefficient was *r* = 0.236, *p* < 0.05. These results indicate a **statistically significant positive correlation**, meaning that as the number of shots increases, home goals tend to increase as well. The correlation is relatively weak, suggesting that other factors also contribute to goal scoring not just shots.


## 6: MACHINE LEARNING

### 6.1 Objective

`Predict total wins per team for the upcoming season`

### 6.2 Aggregating Match Data to Team-Level

**One row per team per season with cumulative stats.**

**Target:** total_wins

**Features:**

    --total_goals, total_goals_conceded
    
    --total_shots, total_shots_on_target
    
    --total_fouls, total_corners
    
    --total_yellow_cards, total_red_cards



```python
# Adding season to the data
data['Season'] = data['Datetime'].dt.year
def get_season(date):
    year = date.year
    if date.month >= 8:  # August+
        return f"{year}/{year+1}"
    else:
        return f"{year-1}/{year}"
        
data['Season'] = data['Datetime'].apply(get_season)
```

```python
# Home stats per team per season
home_stats = data.groupby(['Season', 'HomeTeam']).agg({
    'FTHG':'sum', 'FTAG':'sum', 'HS':'sum', 'HST':'sum',
    'HF':'sum','HC':'sum','HY':'sum','HR':'sum'
}).rename(columns=lambda x: 'home_' + x)

# Away stats per team per season
away_stats = data.groupby(['Season', 'AwayTeam']).agg({
    'FTAG':'sum', 'FTHG':'sum', 'AS':'sum', 'AST':'sum',
    'AF':'sum','AC':'sum','AY':'sum','AR':'sum'
}).rename(columns=lambda x: 'away_' + x)

# Combining home and away stats (one row per team per season)
team_stats = home_stats.add(away_stats, fill_value=0)

# Total features
team_stats['total_goals'] = team_stats['home_FTHG'] + team_stats['away_FTAG']
team_stats['total_goals_conceded'] = team_stats['home_FTAG'] + team_stats['away_FTHG']
team_stats['total_shots'] = team_stats['home_HS'] + team_stats['away_AS']
team_stats['total_shots_on_target'] = team_stats['home_HST'] + team_stats['away_AST']
team_stats['total_fouls'] = team_stats['home_HF'] + team_stats['away_AF']
team_stats['total_corners'] = team_stats['home_HC'] + team_stats['away_AC']
team_stats['total_yellow_cards'] = team_stats['home_HY'] + team_stats['away_AY']
team_stats['total_red_cards'] = team_stats['home_HR'] + team_stats['away_AR']

# Total wins per team per season
team_stats['total_wins'] = (
    data[data['FTR']=='H'].groupby(['Season','HomeTeam']).size().add(
        data[data['FTR']=='A'].groupby(['Season','AwayTeam']).size(), fill_value=0
    )
)

# Reset index for modeling
team_stats = team_stats.reset_index()
team_stats.rename(columns={'HomeTeam': 'Team'}, inplace=True)  # Optional
team_stats.head()
```

### 6.4 Train-Test Split

```python
# Split training and testing seasons
train = team_stats[team_stats['Season'].isin(['2022/2023', '2023/2024'])]
test = team_stats[team_stats['Season'] == '2024/2025']

# Split features and target for train/test
X_train = train.drop(columns=['total_wins', 'Season', 'Team'])
y_train = train['total_wins']

X_test = test.drop(columns=['total_wins', 'Season', 'Team'])
y_test = test['total_wins']
```

### 6.5 Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict wins for test season
y_pred = model.predict(X_test)

# Evaluate
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R²:", r2_score(y_test, y_pred))
```

### 6.7 Predicting Season Winner

```python
# Adding predictions to test DataFrame
test = test.copy()
test['predicted_wins'] = y_pred

# Getting the team name with highest predicted wins
predicted_champion = test.loc[test['predicted_wins'].idxmax(), 'Team']
max_wins = test['predicted_wins'].max()

print(f"Predicted Champion: {predicted_champion}, Wins: {max_wins:.1f}")
```
